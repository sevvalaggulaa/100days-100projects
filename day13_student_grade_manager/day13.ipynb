# Student Grade Manager (Day 13)

```python
#Step 1: Get student scores
student_scores = input("Enter student scores seperated by comas: ")
scores = [int(score.strip()) for score in student_scores.split(",")]

#Step 2: Assign Grades using list Comprehension
grades = [
    "A" if score >= 90 else
    "B" if score >= 80 else
    "C" if score >= 70 else
    "D" if score >= 60 else
    "F" 
    for score in scores
]

#Step 3: Filter Passing and Failing Students
passing_students = [score for score in scores if score >= 60]
failing_students = [score for score in scores if score <60]

# Step 4: Calculate Average Score
average_score = sum(scores) / len(scores)

# Step 5: Sort Scores
sort_choice = input("Sort scores ascending or descending? (asc/desc): ").lower()
if sort_choice == "asc":
    sorted_scores = sorted(scores)
elif sort_choice ==  "desc":
    sorted_scores = sorted(scores, reverse=True)
else:
    sorted_scores = scores
    print("Invalid choice. Scores not sorted. ")

# Step 6: Find Highest and Lowest Scores with Student Details
highest_score = max(scores)
highest_score_student = scores.index(highest_score) + 1

lowest_score = min(scores)
lowest_score_student = scores.index(lowest_score) + 1

#Step 7: Print Results
print("\n--- Student Grades ---")
for i, (score, grade) in enumerate(zip(scores, grades), start=1):
    print(f"Student {i}: Score = {score}, Grade = {grade}")

print("\n--- Passing and Failing Students ---")
print("Passing Students:", passing_students)
print("Failing Students:", failing_students)

print("\n--- Statistics ---")
print(f"Average Score: {average_score:.2f}")
print("Sorted Scores:", sorted_scores)

print("\n--- Highest and Lowest Scores ---")
print(f"Highest Score: Student {highest_score_student}, Score = {highest_score}")
print(f"Lowest Score: Student {lowest_score_student}, Score = {lowest_score}")
```
